```python
import os
import shutil
from pathlib import Path
from transformers import AutoTokenizer, AutoModel, utils
from bertviz.debugger import t5_hook, t5_model_view

utils.logging.set_verbosity_error()  # Suppress standard warnings

debugger_dir = './smdebugger/'
```

```python
model_version = "t5-small" #"gpt2"#'roberta-base' #"gpt2" #"distilbert-base-uncased" #'bert-base-uncased'
model = AutoModel.from_pretrained(model_version)
tokenizer = AutoTokenizer.from_pretrained(model_version)
```

```python
# smdhook = t5_hook(debugger_dir, model, tokenizer)
```

```python
encoder_input_ids = tokenizer("translate English to German: The house is wonderful.", return_tensors="pt", add_special_tokens=True)
with tokenizer.as_target_tokenizer():
    decoder_input_ids = tokenizer("Das Haus ist wunderbar.", return_tensors="pt", add_special_tokens=True)
```

```python
for i in range(2): # run twice to trigger debugger collection
    outputs = model(input_ids = encoder_input_ids.input_ids, decoder_input_ids = decoder_input_ids.input_ids, output_attentions=True)
```
